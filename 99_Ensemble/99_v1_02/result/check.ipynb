```python
import pandas as pd

local_pred_df  = pd.read_csv('stack_99_v1_02_04.csv')
kaggle_pred_df = pd.read_csv('kaggle_stack_df_99_v1_02_04.csv')
```

```python
fold_path = '../../../00_EDA/00_v2_07/result/'

import joblib
print('load folds...')
from os.path import join as opj
val_ids_list = joblib.load(opj(fold_path,f'val_ids_list.joblib'))

fold = 0

local_pred_df = local_pred_df[local_pred_df['essay_id'].isin(val_ids_list[fold])].reset_index(drop=True)
kaggle_pred_df = local_pred_df[['discourse_id']].merge(kaggle_pred_df, on='discourse_id', how='left').reset_index(drop=True)
```

```python
from sklearn.metrics import log_loss

option = ''

local_score = log_loss(local_pred_df['label'].values, local_pred_df[['Ineffective'+option,'Adequate'+option,'Effective'+option]].values)
print('local_score = {:.4f}'.format(local_score))

kaggle_score = log_loss(local_pred_df['label'].values, kaggle_pred_df[['Ineffective'+option,'Adequate'+option,'Effective'+option]].values)
print('kaggle_score = {:.4f}'.format(kaggle_score))
```
