```python
import os
from os.path import join as opj
import pandas as pd

INPUT_PATH = '../../input/asap-aes'

os.listdir(INPUT_PATH)
```

```python
train_df = pd.read_csv(opj(INPUT_PATH,'training_set_rel3.tsv'), sep='\t', encoding='ISO-8859-1')
valid_df = pd.read_csv(opj(INPUT_PATH,'valid_set.tsv'), sep='\t', encoding='ISO-8859-1')
test_df  = pd.read_csv(opj(INPUT_PATH,'test_set.tsv'), sep='\t', encoding='ISO-8859-1')

print('train_df.shape = ', train_df.shape)
print('valid_df.shape = ', valid_df.shape)
print('test_df.shape  = ', test_df.shape)

display(train_df.head(2))
display(valid_df.head(2))
display(test_df.head(2))
```

```python
df = pd.concat([train_df[['essay_id','essay']], 
                valid_df[['essay_id','essay']], 
                test_df[['essay_id','essay']]], axis=0).reset_index(drop=True)
print('df.shape = ', df.shape)
df.head()
```

```python
import os
os.makedirs('result', exist_ok=True)

df.to_csv('result/asap_all_data.csv', index=False)
```
